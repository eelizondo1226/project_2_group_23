```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
# Study data files
campaign_path = "Resources/campaign.csv"
```

```python
# Read the mouse data and the study results
campaign_df = pd.read_csv(campaign_path)
```

```python
campaign_df.head()
```

```python
campaign_df.info()
```

```python
# Filter the DataFrame for backers_count >= 300 and backers_count < 300
filtered_df_high = campaign_df[campaign_df['backers_count'] >= 300]
filtered_df_low = campaign_df[campaign_df['backers_count'] < 300]

# Create a scatter plot to compare backers_count >= 300 and backers_count < 300
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df_high['cf_id'], filtered_df_high['backers_count'], color='blue', label='Backers Count >= 300', s=15)
plt.scatter(filtered_df_low['cf_id'], filtered_df_low['backers_count'], color='red', label='Backers Count < 300', s=15)

plt.xlabel('CF_ID')
plt.ylabel('Backers Count')
plt.title('Comparison of Backers Count >= 300 and < 300 by CF_ID')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

plt.show()
```
