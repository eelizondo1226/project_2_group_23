```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
# Study data files
campaign_path = "Resources/campaign.csv"
```

```python
# Read the mouse data and the study results
campaign_df = pd.read_csv(campaign_path)
```

```python
campaign_df.head()
```

```python
campaign_df.info()
```

```python
# Filter the DataFrame for backers_count >= 300 and backers_count < 300
filtered_df_high = campaign_df[campaign_df['backers_count'] >= 300]
filtered_df_low = campaign_df[campaign_df['backers_count'] < 300]

# Create a scatter plot to compare backers_count >= 300 and backers_count < 300
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df_high['cf_id'], filtered_df_high['backers_count'], color='blue', label='Backers Count >= 300', s=15)
plt.scatter(filtered_df_low['cf_id'], filtered_df_low['backers_count'], color='red', label='Backers Count < 300', s=15)

plt.xlabel('CF_ID')
plt.ylabel('Backers Count')
plt.title('Comparison of Backers Count >= 300 and < 300 by CF_ID')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

plt.show()
```

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import seaborn as sns
from scipy.stats import linregress

# Study data files
campaign = "Resources/campaign.csv"
subcat = "Resources/subcategory.csv"

# Read in csv files
campaign_data = pd.read_csv(campaign)
subcat_data = pd.read_csv(subcat)

# Combine the data into a single DataFrame
merged_df = pd.merge(campaign_data, subcat_data, how="left", on=["subcategory_id", "subcategory_id"])

# Display the dataframe
merged_df.head()
```

```python
merged_df_sorted = merged_df.sort_values(ascending=False, by='pledged')
merged_df_sorted.head()
```

```python
x = merged_df_sorted['subcategory']
y = merged_df_sorted['pledged']


plt.figure(figsize=(8,6))
plt.barh(x, y, color = "#1a759f")

# Plot customization
plt.ylabel("Subcategory", fontsize=12, fontstyle="italic")
plt.xlabel("Pledged (U.S. Dollars)", fontsize=12, fontstyle="italic")
plt.title("Subcategory Ranked By Pledged Amound", fontsize=16)
plt.grid(axis="x", color="lightgrey", linestyle="--", alpha=0.7)

plt.show()
```
